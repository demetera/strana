```python
import pandas as pd
```

```python
data = pd.read_csv('../data/strana_articles.csv')
data_content = data['content']
```

```python
data_sample = data_content[:10]
```

```python
# processing articles (removal of links, extra etc.) without split
# without splitting to sentences

def remove_links(s):
    while True:
        ind1 = s.find('t.me')
        if ind1 != -1:
            ind2 = s[ind1:].find(' ')
            s = s.replace(s[ind1 : ind1 + ind2],' ')
        else:
            break
    while True:
        ind1 = s.find('Фото: ')
        if ind1 != -1:
            ind2 = s[ind1 + 7:].find(' ')
            s = s.replace(s[ind1 : ind1 + 7 + ind2],' ')
        else:
            break
    while True:
        ind1 = s.find('Видео: ')
        if ind1 != -1:
            ind2 = s[ind1 + 8:].find(' ')
            s = s.replace(s[ind1 : ind1 + 8 + ind2],' ')
        else:
            break
    return s

def process_content(art: str):
    art = art.replace('\xa0', ' ').replace('[...]', ' ')
    art = art.replace('Прим.Ред','').replace('Прим. ред.)','')
    art = art.replace(' п.',' п').replace(' ч.', ' ч').replace(' ст.',' ст').replace(' им.', ' им').replace('г. ', 'г')
    art = art.replace('прим.ред','').replace('ред.','ред ').replace('ул.','ул ').replace('..', ' ').replace('...', ' ').replace('"', '')
    art = art.replace('и.о.','ио').replace('и. о.','ио')
    # art = art.replace('-', ' ').replace('–', ' ') # different dashes
    art = remove_links(art)
    words = art.split() # REMOVES
    art = ' '.join(words) # EXTRA SPACES
    return art
```

```python
data_mod = data_sample.map(process_content)
```
